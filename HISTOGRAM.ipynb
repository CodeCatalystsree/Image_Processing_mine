```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


```

```python
#histogram of image
img=cv2.imread('IPPO.JPEG')
hist=cv2.calcHist([img],[0],None,[256],[0,256])
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("GREYSCALE IMAGE")
plt.axis('off')
plt.subplot(1,2,2)
plt.plot(hist,color='black')
plt.title("HISTOGRAM")
plt.xlabel("pixel intensity")
plt.ylabel("no. of pixel")
plt.show()
```

```python
#histogram of image IN RGB PLANE
img=cv2.imread('IPPO.JPEG')
colors=('blue','green','red')
plt.figure(figsize=(8,5))
for i,color in enumerate(colors):
    hist=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=color)
'''plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("GREYSCALE IMAGE")
plt.axis('off')'''
plt.title("RGB HISTOGRAM")
plt.xlabel("pixel intensity")
plt.ylabel("no. of pixel")
plt.show()
```

```python
#histogram equalization
img=cv2.imread('IPPO2.JPEG',0)
img=np.uint8(img)
equa_img=cv2.equalizeHist(img)

equa_hist=cv2.calcHist([equa_img],[0],None,[256],[0,256])
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.title("image")
plt.subplot(1,3,2)
plt.imshow(equa_img,cmap='gray')
plt.title("equalized image")

plt.subplot(1,3,3)
plt.plot(equa_hist)
plt.title("equalized histogram")
plt.show()
```

```python
#histogram streatching or contrast streaching
img=cv2.imread('IPPO.JPEG')
p_min=np.min(img)
p_max=np.max(img)

streatched_img=((img-p_min)/(p_max-p_min))*255
streatched_img=np.uint(streatched_img)

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(streatched_img,cmap='gray')
```

```python
#histogram streatching or contrast streaching
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('scene.JPEG',0)
p_min=np.min(img)
p_max=np.max(img)

streatched_img=((img-p_min)/(p_max-p_min))*255
streatched_img=np.uint(streatched_img)

plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(2,2,2)
plt.hist(img.ravel(),bins=256,range=[0,256])

plt.subplot(2,2,3)
plt.imshow(streatched_img,cmap='gray')

plt.subplot(2,2,4)
plt.hist(streatched_img.ravel(),bins=256,range=[0,256])

plt.show()
```
