```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#gaussian noise
img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
org_img=img.copy()
mean=float(input("enter the mean value"))
sigma=float(input("enter the sigma value"))
def add_gauss_noise(image,mean,sigma):
    gauss=np.random.normal(mean,sigma,image.shape).astype(np.uint8)
    noisy=cv2.add(image,gauss)
    return noisy

gauss_noisy_img=add_gauss_noise(img,mean,sigma)

plt.subplot(1,2,1)
plt.imshow(org_img,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(gauss_noisy_img,cmap='gray')
plt.show()
```

```python
#salt and pepper noise
img=cv2.imread('IPPO2.JPEG',cv2.IMREAD_GRAYSCALE)
org_img=img.copy()
s_p=float(input("enter the salt prob"))
p_p=float(input("enter the pepper prob"))

def salt_pepp_noise(image,s_p,p_p):
    total_pixel=image.size
    num_salt=int(s_p*total_pixel)
    salt_cords=[np.random.randint(0,i,num_salt) for i in image.shape]
    org_img[salt_cords[0],salt_cords[1]]=255

    num_pepper=int(p_p*total_pixel)
    pepp_cords=[np.random.randint(0,i,num_pepper) for i in image.shape]
    org_img[pepp_cords[0],pepp_cords[1]]=0
    return org_img
noisy_img=salt_pepp_noise(img,s_p,p_p)

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(noisy_img,cmap='gray')

plt.show()
```
