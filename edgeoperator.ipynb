```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#sobel operator
image=cv2.imread('IPPO2.JPEG',0)
sobel_x=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
soble_y=cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)
soble_combined=cv2.magnitude(sobel_x,soble_y)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(sobel_x,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(soble_y,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(soble_combined,cmap='gray')

plt.show()
```

```python
#prewit operator
img=cv2.imread('IPPO.JPEG',0)
kernel_x=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
kernel_y=np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
pre_x=cv2.filter2D(img,-1,kernel_x)
pre_y=cv2.filter2D(img,-1,kernel_y)
pre_img=cv2.magnitude(pre_x.astype(np.float32),pre_y.astype(np.float32))
plt.imshow(pre_img,cmap='gray')
plt.show()
```

```python
#robert operator
img=cv2.imread('IPPO.JPEG',0)
kernel_x=np.array([[1,0],[0,-1]])
kernel_y=np.array([[0,1],[-1,0]])
r_x=cv2.filter2D(img,-1,kernel_x)
r_y=cv2.filter2D(img,-1,kernel_y)
robert_img=cv2.magnitude(r_x.astype(np.float32),r_y.astype(np.float32))
plt.imshow(robert_img,cmap='gray')
plt.show()
```
