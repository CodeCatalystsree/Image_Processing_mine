```python
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('IPPO.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
#plt.axis('off')
plt.show()
```

```python
img=cv2.imread('IPPO.JPEG')
img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img)
plt.show()
```

```python
img=cv2.imread('IPPO.JPEG')
img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
gray_img=img
_, bin_img=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
plt.imshow(bin_img)
plt.show()
#rgb to gray to binary because direct rgb to binary not possible
```

```python
gray_img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
_, binary_img=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
plt.imshow(binary_img,cmap='gray')
plt.axis('off')
plt.show()
#gray to binary is done by thresholding
```

```python
img=cv2.imread('IPPO.JPEG')
hsv_img=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
plt.show()
#rgb to hsv
```

```python
img=cv2.imread('IPPO.JPEG')
RGB_img=cv2.cvtColor(img,cv2.COLOR_HSV2RGB)
plt.imshow(RGB_img)
plt.show()
#hsv to rgb
```

```python
img=cv2.imread('IPPO.JPEG')
rgb_img=cv2.cvtColor(img,cv2.COLOR_HSV2RGB)
gray_img=cv2.cvtColor(rgb_img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img)
plt.show()
#hsv to rgb to gray direct hsv to gray not possible
```

```python
gray_img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
RGB_img=cv2.cvtColor(gray_img,cv2.COLOR_GRAY2RGB)
plt.imshow(RGB_img)
plt.show()
#graytorgb
```

```python
gray_img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
RGB_img=cv2.cvtColor(gray_img,cv2.COLOR_GRAY2RGB)
hsv_img=cv2.cvtColor(RGB_img,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
plt.show()
#graytorgbtohsv because direct conversion from gray to hsv not possible
```

```python
gray_img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
RGB_img=cv2.cvtColor(gray_img,cv2.COLOR_GRAY2RGB)
hsv_img=cv2.cvtColor(RGB_img,cv2.COLOR_RGB2HSV)

plt.subplot(1,2,1)
plt.imshow(hsv_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(gray_img)
plt.axis('off')
plt.show()
#graytorgbtohsv because direct conversion from gray to hsv not possible using subplot i can show input out image side by side
```

```python
#IMAGE ADDITION-BRIGHTNESS INCREASE
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('IPPO.JPEG')
bright_img=cv2.add(img,np.array([50.0]))
plt.imshow(cv2.cvtColor(bright_img,cv2.COLOR_BGR2RGB))
plt.show()

```

```python
#IMAGE ADDITION-BRIGHTNESS INCREASE
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('IPPO.JPEG')
dark_img=cv2.substract(img.shape,np.full(50.0,dtype=np.unit8))
plt.imshow(cv2.cvtColor(dark_img,cv2.COLOR_BGR2RGB))
plt.show()

```
