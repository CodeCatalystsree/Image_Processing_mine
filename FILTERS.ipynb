```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#average or mean filter
img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
k_size=(3,3)
filtered_img=cv2.blur(img,k_size)
plt.imshow(filtered_img,cmap='gray')
plt.show()
```

```python
#max filter
img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
k_size=np.ones((3,3),np.uint8)
filtered_img=cv2.dilate(img,k_size)
plt.imshow(filtered_img,cmap='gray')
plt.show()
```

```python
#min filter
img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
k_size=np.ones((3,3),np.uint8)
filtered_img=cv2.erode(img,k_size)
plt.imshow(filtered_img,cmap='gray')
plt.show()
```

```python
#median filter
img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
#k_size=np.ones((3,3),np.uint8)
filtered_img=cv2.medianBlur(img,3)
plt.imshow(filtered_img,cmap='gray')
plt.show()
```

```python
#gaussian filter
img=cv2.imread('IPPO.JPEG',cv2.IMREAD_GRAYSCALE)
k_size=np.ones((3,3),np.uint8)
filtered_img=cv2.GaussianBlur(img,(3,3),5)
plt.imshow(filtered_img,cmap='gray')
plt.show()
```
