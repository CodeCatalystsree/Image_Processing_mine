```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#bitplane slicing

img=cv2.imread('IPPO2.JPEG',0)
h,w=img.shape

bit_planes=[]
for i in range(80):
    bit_plane=(img>>i)*1
    bit_plane=bit_plane*255
    bit_planes.append(bit_plane)
plt.figure(figsize=(12,5))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(bit_planes[i],cmap='gray')

plt.show()
```

```python
#gray level slicing without background
img=cv2.imread('IPPO.JPEG',0)

min_int=int(input("enter min intensity"))
max_int=int(input("enter the max intensity"))
sliced_img=np.where((img>=min_int)&(img<=max_int),255,0).astype(np.uint8)
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(sliced_img,cmap='gray')
plt.show()


```

```python
#gray level slicing with background
img=cv2.imread('IPPO.JPEG',0)

min_int=int(input("enter min intensity"))
max_int=int(input("enter the max intensity"))
sliced_img=np.where((img>=min_int)&(img<=max_int),255,img).astype(np.uint8)
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(sliced_img,cmap='gray')
plt.show()


```
